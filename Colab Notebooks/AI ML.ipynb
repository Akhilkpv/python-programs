```python
from datetime import datetime
import pytz
def chatbot_response(user_input):
  user_input = user_input.lower()# Convert input into lowercase

  #Rule-based responces
  if "hello" in user_input or "hi" in user_input:
    return "Hello! How can I assist you today?"
  elif "how are you" in user_input:
    return "I'm just a bot, but I'm doing great! How are you? "
  elif "time" in user_input:
    indian_timezone = pytz.timezone("Asia/Kolkata")
    current_time = datetime.now(indian_timezone).strftime("%H:%M:%S")
    return f"The current time is {current_time}."
  elif "bye" in user_input or "goodbye" in user_input:
    return "Goodbye! Have a great day!"
  else:
    return "I'm sorry, I didn't understand that. Can you please rephrase?"
#main loop for the chatbot
def chat():
  print("Chatbot:Hello! I'm a simple chatbot. Type 'bye' or 'exit' to end the conversation.")
  while True:
    user_input = input("You: ")
    response = chatbot_response(user_input)
    print(f"Chatbot:, {response}")
    if "bye" in user_input or "exit" in user_input:
      break
# Run the bot
chat()
```

```python
from textblob import TextBlob
def analyze_sentiment(text):
  #create a textblob object
  blob = TextBlob(text)
  # got the sentiment polarity (-1 to 1)
  sentiment = blob.sentiment.polarity
  if sentiment > 0:
    return "Positive"
  elif sentiment < 0:
    return "Negative"
  else:
    return "Neutral"
# Example usage
user_input = input("Enter a sentence to analyse sentiment : ")
sentiment = analyze_sentiment(user_input)
print(f"The sentiment of the input is: {sentiment}")
```

```python
!pip install gradio transformers
import gradio as gr
from transformers import pipeline

#Load the question answering mode
qa_pipeline = pipeline("question-answering",model="distilbert-base-uncased-distilled-squad")
#Define the Gradio interface
def answer_question(context,question):
  result = qa_pipeline(context=context,question=question)
  return result["answer"]
#Create the Gradio interface

interface = gr.Interface(fn=answer_question,
                         inputs=["text","text"],
                         outputs="text",
                         live=True)
interface.launch()
```
